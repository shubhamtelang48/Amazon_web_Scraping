```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
url="https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1"
HEADERS = ({'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',
                                'Accept-Language': 'en-US, en;q=0.5'})
webpage=requests.get(url, headers=HEADERS)
```

```python
soup1 = BeautifulSoup(webpage.content, "html.parser")

soup2 = BeautifulSoup(soup1.prettify(), "html.parser")
```

```python
print(soup2)
```

```python

box1=soup2.find("div",class_="s-main-slot s-result-list s-search-results sg-row")
print(box1)
```

```python
names =box1.find_all("h2", class_="a-size-mini a-spacing-none a-color-base s-line-clamp-2")
Product_name=[]
for i in names:
    name=i.text.strip()
    Product_name.append(name)
print(Product_name)
print(len(Product_name))
```

```python
price = box1.find_all("span", class_="a-price-whole")
Product_price=[]
for i in price:
    name=i.text.strip()
    Product_price.append(name)
print(Product_price)
print(len(Product_price))
```

```python
Product_price=Product_price[0:21]
len(Product_price)
```

```python
reviews = box1.find_all("span",class_="a-size-base s-underline-text")
Number_of_Review=[]
for i in reviews:
    name=i.text.strip()
    Number_of_Review.append(name)
print(Number_of_Review)
print(len(Number_of_Review))
```

```python
rating = box1.find_all("span", class_="a-icon-alt")
Rating=[]
for i in rating:
    name=i.text.strip()
    Rating.append(name)
print(Rating)
print(len(Rating))
```

```python
url = soup2.find("a", class_="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal").get("href")
```
